# Demonstration of empirical distribution

Generate `N` random points in a rectangle. Count the number of points that are in the first, second and the third parts of the rectangle.


```python
import matplotlib.pyplot as plt
from numpy import random

plt.plot([0, 3, 3, 0, 0], [0, 0, 1, 1, 0], color = 'black')
plt.plot([1,1],[0,1],color='black')
plt.plot([2,2],[0,1],color='black')

N = 100   # number of particles
xs = [random.rand()*3 for _ in range(N)]
ys = [random.rand() for _ in range(N)]
plt.scatter(xs,ys, c=xs, cmap='cool')
plt.show()
```

```python
# count the number of points in each parts
n1 = sum([1 if x<=1 else 0 for x in xs]) 
n2 = sum([1 if x> 1 and x<=2 else 0 for x in xs]) 
n3 = sum([1 if x>=2 else 0 for x in xs]) 
```

```python
print(n1,n2,n3)
```

```python
print(f"empirical distribution is ({n1/N},{n2/N}, {n3/N})")
```
